## **Model Building**

### Loading the model

```python
model_name_or_path = "TheBloke/Llama-2-13B-chat-GGUF" #Llama 2 with 13B parameters
model_basename = "llama-2-13b-chat.Q5_K_M.gguf" # the model is in gguf format - quantized with 5 bit
```

```python
# Using hf_hub_download to download a model from the Hugging Face model hub
# The repo_id parameter specifies the model name or path in the Hugging Face repository
# The filename parameter specifies the name of the file to download

model_path = hf_hub_download(
    repo_id = model_name_or_path,    # complete the code to mentioned the repo_id
    filename = model_basename    # complete the code to mentioned the filename
)
```

```python
# Create an instance of the 'Llama' class with specified parameters

lcpp_llm = Llama(
    model_path=model_path,
    n_threads=None,   # CPU cores
    n_batch=1024,      # Should be between 1 and n_ctx, consider the amount of VRAM in your GPU.
    n_gpu_layers=-1,  # Change this value based on your model and your GPU VRAM pool.
    n_ctx=1024,        # Context window
)
```

### Defining Model Response Parameters

```python
def generate_llama_response(row, total, instruction):

    review = row['support_ticket_text']
    current = row.name + 1

    # System message explicitly instructing not to include the review text
    system_message = """
        [INST]<<SYS>>
        {}
        <</SYS>>[/INST]
    """.format(instruction)

    # Combine user_prompt and system_message to create the prompt
    prompt = f"{review}\n{system_message}"

    print(f'\n=> Generating response for row {current} of {total} >>>>')
    # Generate a response from the LLaMA model
    response = lcpp_llm(
        prompt=prompt,
        max_tokens=1024,
        temperature=0.01,
        top_p=0.95,
        repeat_penalty=1.2,
        top_k=50,
        stop=['INST'],
        echo=False,
        seed=42,
    )

    # Extract the sentiment from the response
    response_text = ""
    for choice in response["choices"]:
      response_text = response_text + ' ' + choice["text"]
    return response_text
```

## **Task 1: Ticket Categorization**

```python
# create a copy of the data
data_1 = data.copy()
```

```python
## Complete the prompt to get the desired output
instruction_1 = """
    You are a customer support executive for computer related problem. Your job is to create ticket categories based on the problem given by the user.
    The categories are : 'technical issues', 'hardware issues' and 'data recovery'.
    Please answer only with one possible category name.
    Do not include any other text other than the category text.
"""
```

```python
data_1['llama_response'] = data_1.apply(generate_llama_response, axis=1, args=(len(data_1), instruction_1))
```

```python
# Write the code to check the first five rows of the data to confirm whether the new column has been added
data_1.head(5)
```

```python
def extract_category(model_response):
    if 'technical issues' in model_response.lower():
        return 'Technical issues'
    elif 'hardware issues' in model_response.lower():
        return 'Hardware issues'
    elif 'data recovery' in model_response.lower():
        return 'Data recovery'
```

```python
data_1['Category'] = data_1['llama_response'].apply(extract_category)
data_1['Category'].head()
```

```python
#Drop the llama response column
final_data_1 = data_1.drop(['llama_response'], axis=1)
final_data_1.head()
```

## **Task 2: Ticket Categorization and Returning Structured Output**

```python
# create a copy of the data
data_2 = data.copy()
```

```python
## Complete the prompt to get the desired output
instruction_2 = """
    You are a customer support executive for computer related problem.
    Your job is to create ticket categories based on the problem given by the user.
    The categories are : 'technical issues', 'hardware issues' and 'data recovery'.
    Please answer using following json structure:
    { 'Category' : '' }
    Do not include any other text in the output except the JSON.
"""
```

```python
data_2['llama_response'] = data_2.apply(generate_llama_response, axis=1, args=(len(data_2), instruction_2))
```

```python
# Write the code to check the first five rows of the data to confirm whether the new column has been added
data_2.head(5)
```

```python
# defining a function to parse the JSON output from the model
def extract_json_data(json_str):
    try:
        # Find the indices of the opening and closing curly braces
        json_start = json_str.find('{')
        json_end = json_str.rfind('}')

        if json_start != -1 and json_end != -1:
            extracted_category = json_str[json_start:json_end + 1]  # Extract the JSON object
            data_dict = json.loads(extracted_category)
            return data_dict
        else:
            print(f"Warning: JSON object not found in response: {json_str}")
            return {}
    except json.JSONDecodeError as e:
        print(f"Error parsing JSON: {e}")
        return {}

#Define a function to convert a string to title case
def to_title_case(str):
  return str.title()
```

```python
#Extract the json body
data_2['llama_response_parsed'] = data_2['llama_response'].apply(extract_json_data)
data_2['llama_response_parsed'].head()
```

```python
#normalize json into panda column
llama_response_parsed_df_2 = pd.json_normalize(data_2['llama_response_parsed'])
#Convert the Category values to title case
llama_response_parsed_df_2['Category'] = llama_response_parsed_df_2['Category'].apply(to_title_case)
#Show first 5 rows
llama_response_parsed_df_2.head()
```

```python
#Add the extracted json column to original dataset
data_with_parsed_model_output_2 = pd.concat([data_2, llama_response_parsed_df_2], axis=1)
data_with_parsed_model_output_2.head()
```

```python
#Drop the intermediate columns
final_data_2 = data_with_parsed_model_output_2.drop(['llama_response','llama_response_parsed'], axis=1)
final_data_2.head()
```
